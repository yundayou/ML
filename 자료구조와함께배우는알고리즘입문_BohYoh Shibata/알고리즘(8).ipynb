```python
pos = [0] * 8
flag = [False] * 8

def put() ->None:
    for i in range(8):
        print(f'{pos[i]:2}', end='')
    print()
    
def set(i : int)->None:
    for j in range(8):
        if not flag[j]:
            pos[i] = j
            if i == 7:
                put()
            else:
                flag[j] = True
                set(i+1)
                flag[j] = False
set(0)
```

```python
pos = [0] * 8
flag_a = [False] * 8
flag_b = [False] * 15
flag_c = [False] * 15

def put() -> None:
    for j in range(8):
        for i in range(8):
            print('0' if pos[i] == j else 'ㅁ', end = '')
        print()
    print()
    
def set(i : int) -> None:
    for j in range(8):
        if(    not flag_a[j]
           and not flag_b[j+i]
           and not flag_c[i - j +7]):
            pos[i] = j
            if i == 7:
                put()
            else:
                flag_a[j] = flag_b[i + j] = flag_c[i - j +7] = True
                set(i+1)
                flag_a[j] = flag_b[i + j] = flag_c[i - j + 7] = False
                
set(0)
```
