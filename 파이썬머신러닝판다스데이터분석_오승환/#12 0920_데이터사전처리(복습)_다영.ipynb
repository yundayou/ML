### 데이터 사전처리
#### 데이터 분석의 정확도 => 데이터 품질에 의해 좌우
#### 누락데이터, 중복데이터 등 오류를 수정, 분석목적에 맞게 변형하는 과정이 매우 중요

* 누락데이터
>-  value_counts() 메서드 이용, dropna = False
>-  isnull() : 누락데이터 -> True, 유효한 데이터 -> False
>-  notnull() : 누락데이터 -> False, 유효한 데이터 -> True
>-  dropna(axis = , thresh=, how ='any'(nan값이 하나라도 존재하면 삭제), how = 'all'(모든 데이터가 nan 값이어야만 삭제)
>-  fillna(mean_age, inplace =True) : 누락데이터 치환, method = 'ffill' : 직전 행의 값으로 변경, mehtod = 'bfill' : 다음 행의 값으로 변경



* 중복데이터
>- duplicated() :  각 행의 중복여부를 나타내는 불린 시리즈를 반환
>- drop_duplicates(subset = '열 이름의 리스트') 

* 데이터 표준화
>- 단위환산
>- 자료형 변환 :  유한 개의 고유값이 반복적으로 나타나는 경우에는 범주형 데이터로 표현하는 것이 효율적
>- 구간분할 : 일정한 구간으로 나눠서 분석하는 것이 효율적 (np.histogram(data , bins = n)
>- 더미변수 : 0과 1로 표현, 수의 크고 작음을 나타내지 않고, 어떤 특성이 있는 지 없는 지의 여부만 표시(원핫벡터, 원핫코딩) cf) 희소행렬 : 행렬의 대부분이 0으로 구성




* 정규화
>- 숫자 데이터의 상대적인 크기 차이를 제거, 각 열에 속하는 데이터 값을 동일한 크기 기준으로 나눈 비율로 나타내는 것
>- 최대값으로 나누는 방법 -> 가장 큰 값 :1
>- (최대값 - 최소값)으로 나누기 : 0 ~ 1 사이에 값 포진


* 시계열 데이터
>- Timestamp : 특정한 시점을 기록
>- Timestamp 배열 :  time_range(start, end, periods, freq, tz) 함수 이용
>- Period : 두 시점 사이의 일정한 기간을 나타냄
>- freq 옵션 : 'D' : 1일, 'M' : 1달, 'A' :  1년
>- to_datetime() : 다른 자료형 -> Timestamp(datetime64)로 변환


* 시계열 데이터 활용
>- 날짜 데이터 분리 : dt.year, dt.month, dt.day로 추출
>- 날짜 인덱스 사용 장점 :  연-월-일 중에서 내가 필요로 하는 레벨을 선택적으로 인덱싱 가능 , 날짜 범위로 슬라이싱 추출도 가능
