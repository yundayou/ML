* 함수매핑
>- 시리즈 원소에 함수매핑(apply()) : 인자로 전달하는 매핑 함수에 시리즈의 모든 원소를 하나씩 입력하고 함수의 리턴값을 돌려받는다. Series객체.apply(매핑함수)
>- 데이터프레임이 원소에 함수매핑(applymap(매핑함수)) : 매핑 함수에 데이터프레임의 각 원소를 하나씩 넣어서 리턴값으로 돌려받는다. 
>- DataFrame.apply(매핑함수, axis = 0)
>- 데이터프레임 객체에 함수 매핑 ((DataFrame객체.pipe(매핑함수)) : 사용하는 함수가 반환하는 리턴값에 따라 pipe() 메소드가 반환하는 객체의 종류가 결정

* 열재구성
>- 열 순서변경 : DataFra me 객체[재구성한 열 이름의 리스트]

* 필터링
>- 시리즈 또는 데이터프레임의 데이터 중에서 특정 조건식을 만족하는 원소만 따로 추출하는 개념 
>- isin() : 특정 값을 가진 행들을 따로 추출 가능, DataFrame의 열객체.isin(추출값의리스트)

* 데이터프레임 합치기
>- 데이터 프레임 concat/merge/join
>- >pandas.concat(데이터프레임리스트), join = 'outer'이 default: 행인덱스의 합집합, join = 'inner' :  행인덱스의 교집합, ignore_index = True :  새로운 인덱스 설정, 
>- >pandas.merge(df_left, df_right, how = 'inner', on = None), 키가 되는 열이나 인덱스가 반드시 양쪽 데이터프레임에 모두 존재, how = 'outer' 한쪽에만 존재하더라도 포함한다
>- >DataFrame1.join(DataFrame2, how = 'left') :  merger와 거의 흡사




* 그룹연산(SQL과 비슷)

>- 그룹연산 : DataFrame객체.groupby(기준이 되는 열)
>- > 여러열을 기준: DataFrame객체.groupby(기준이 되는 열의 리스트)
>- 특정그룹만을 골라서 추출 :  get_group(('Third', 'female'))
>- 그룹 연산 메서드 : group객체.std(), agg()메소드 집계 : group객체.agg(매핑함수)
>- 모든 열에 여러함수를 매핑 : group객체.agg([함수1, 함수2, 함수3...])
>- 각 열마다 다른 함수를 매핑 : group객체.agg({'열1' : 함수1, '열2' : 함수2, ...})
>- 그룹 연산 데이터 변환 : group객체.transform(매핑함수)
>- > transform() : 그룹별로 구분하여 각 원소에 함수를 적용, 그룹별 집계 대신 각 원소의 본래 행 인덱스와 열 이름을 기준으로 연산 결과를 반환한다. 
>- 그룹객체 필터링 : group객체.filter(조건식 함수)
